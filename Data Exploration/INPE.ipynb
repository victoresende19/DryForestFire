```python
import pandas as pd
import numpy as np
```

```python
# Carrega o arquivo CSV em um DataFrame
df = pd.read_csv('historico_estado_distrito_federal.csv', sep=',')
df = df.drop(columns=df.columns[-1])

# Transforma o DataFrame em um formato "tidy"
df = df.melt(id_vars=['Unnamed: 0'], var_name='mes', value_name='frequencia')

# Renomeia a coluna "Unnamed: 0" para "ano"
df = df.rename(columns={'Unnamed: 0': 'ano'})

df.frequencia = df.frequencia.replace('-', 0).astype('int64')

# Exibe o DataFrame resultante
df.sort_values(by = ['ano']).reset_index(drop=True)
```

```python
df['month'] = df['mes'].replace({
        'Janeiro': 1,
        'Fevereiro': 2,
        'Março': 3,
        'Abril':4,
         'Maio':5,
         'Junho':6,
         'Julho':7,
         'Agosto':8,
         'Setembro':9,
         'Outubro':10,
         'Novembro':11,
         'Dezembro':12
        }, regex=True)
```

```python
df['month'] = df['month'].astype(str)
df['ano'] = df['ano'].astype(str)
```

```python
df['Date'] = pd.to_datetime(df["month"].str.cat(df["ano"], sep = "/"), format='%m/%Y')
df['Date'] = df['Date'].dt.strftime('%m/%Y')
```
